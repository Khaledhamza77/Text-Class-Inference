# Dataset Reading and Embedding

https://huggingface.co/sentence-transformers/all-MiniLM-L6-v2

```python
model=SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')
```

```python
df=pd.read_excel('/content/Generated Tweets.xlsx')
df
```

```python
embeddings=model.encode(df['text'].to_list())
embed_list=[]
for i in range(df.shape[0]):
    x=[]
    for j in range(384):
        x.append(embeddings[i][j])
    embed_list.append(x)
```

```python
df['embedding']=embed_list
```

```python
# Expand lists into separate columns
expanded_df = df['embedding'].apply(pd.Series)

# Concatenate the expanded columns with the original DataFrame
result_df = pd.concat([df, expanded_df], axis=1).drop('embedding', axis=1)

# Display the result
result_df
```

```python
X=result_df.copy()
X.drop(['text','label'],axis=1,inplace=True)
X
```

# PCA Decomposition

```python
n_comp=min(X.shape[0],X.shape[1])
pca = PCA(n_components=n_comp)
components=pca.fit_transform(X)
pca_varPerc=np.round(pca.explained_variance_ratio_* 100,decimals=1)
pca_varPerc
```

```python
n=0
for i in range(n_comp):
  if np.round(sum(pca.explained_variance_ratio_[0:i]* 100),decimals=0) == 100:
    n=i
    break
n
```

```python
PCA1=pd.DataFrame([])
for i in range(n):
  PCA1[str(i+1)]=components[:,i]
PCA1
```

# Number of Clusters Inference (with full embedding): Silhouette Score

```python
silh=pd.DataFrame([])
for i in range(2,21):
  clust=SpectralClustering(n_clusters=i,random_state=42)
  clust.fit(X)
  score=silhouette_score(X,clust.labels_)

  j=silh.shape[0]
  silh.at[j,'k']=i
  silh.at[j,'sil']=score

# Plot the elbow curve
plt.plot(range(2, 21), silh['sil'], marker='o')
plt.title('Elbow Method for Optimal k')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Scores')
plt.show()
```

```python
best_k=int(silh.at[np.argmax(silh['sil']),'k'])
best_k
```

# Clustering (Full Embedding)

```python
spectclus=SpectralClustering(n_clusters=best_k,random_state=42)
spectclus.fit(X)
spectclus.labels_
```

# Clustering (Dimesionality Reduced)

## Model - DistilBert

https://huggingface.co/ml6team/keyphrase-extraction-distilbert-openkp

```python
# Define keyphrase extraction pipeline
class KeyphraseExtractionPipeline(TokenClassificationPipeline):
    def __init__(self, model, *args, **kwargs):
        super().__init__(
            model=AutoModelForTokenClassification.from_pretrained(model),
            tokenizer=AutoTokenizer.from_pretrained(model),
            *args,
            **kwargs
        )

    def postprocess(self, all_outputs):
        results = super().postprocess(
            all_outputs=all_outputs,
            aggregation_strategy=AggregationStrategy.FIRST,
        )
        return np.unique([result.get("word").strip() for result in results])
```

```python
model_name = "ml6team/keyphrase-extraction-distilbert-openkp"
extractor = KeyphraseExtractionPipeline(model=model_name)
```

```python
from transformers import pipeline
unmasker = pipeline('fill-mask', model='bert-large-uncased')
```

```python
def get_embeddings(df_path):
  df=pd.read_excel(df_path)
  embeddings=model.encode(df['text'].to_list())
  embed_list=[]
  for i in range(df.shape[0]):
      x=[]
      for j in range(384):
          x.append(embeddings[i][j])
      embed_list.append(x)
  df['embedding']=embed_list
  expanded_df = df['embedding'].apply(pd.Series)
  result_df = pd.concat([df, expanded_df], axis=1).drop('embedding', axis=1)
  X=result_df.copy()
  X.drop(['text','label'],axis=1,inplace=True)
  return df, X
```

```python
def plot(X):
  n_comp=min(X.shape[0],X.shape[1])
  pca = PCA(n_components=n_comp)
  components=pca.fit_transform(X)
  pca_varPerc=np.round(pca.explained_variance_ratio_* 100,decimals=1)
  n=0
  for i in range(n_comp):
    if np.round(sum(pca.explained_variance_ratio_[0:i]* 100),decimals=0) == 100:
      n=i
      break
  PCA1=pd.DataFrame([])
  for i in range(n):
    PCA1[str(i+1)]=components[:,i]
  PCA1
  silh=pd.DataFrame([])
  for i in range(2,21):
    clust=SpectralClustering(n_clusters=i,random_state=42)
    clust.fit(X)
    score=silhouette_score(X,clust.labels_)

    j=silh.shape[0]
    silh.at[j,'k']=i
    silh.at[j,'sil']=score

  # Plot the elbow curve
  plt.plot(range(2, 21), silh['sil'], marker='o')
  plt.title('Elbow Method for Optimal k')
  plt.xlabel('Number of Clusters (k)')
  plt.ylabel('Silhouette Scores')
  plt.savefig('plot_silh.png')
  plt.close()
  silh_image = Image.open('plot_silh.png')

  best_k=int(silh.at[np.argmax(silh['sil']),'k'])
  spectclus=SpectralClustering(n_clusters=best_k,random_state=42)
  spectclus.fit(PCA1)
  labels=spectclus.labels_

  plt.scatter(PCA1['1'],PCA1['2'],c=spectclus.labels_, cmap='viridis', s=50)
  plt.title(f'Scatter Plot of Data (Spectral)')
  plt.xlabel('PC1')
  plt.ylabel('PC2')
  plt.colorbar(label='Cluster Assignment')
  plt.savefig('plot_clust.png')
  clust_image = Image.open('plot_clust.png')

  return labels, best_k, silh_image, clust_image
```

```python
def main_function(df_path):
  df, X=get_embeddings(df_path)
  labels, best_k, silhplot, clustplot = plot(X)

  data=df.copy()
  data['cluster']=labels
  data.drop('embedding',axis=1,inplace=True)
  data

  summaries=[]
  for i in range(best_k):
    text=''
    c=data[data['cluster']==i]
    c.reset_index(inplace=True,drop=True)

    for j in range(c.shape[0]):
      text+=c.at[j,'text']+' '

    summarized = summarizer(text)
    summaries.append(summarized[0]['summary_text'])

  sum = ""
  clas = ""
  for i in range(len(summaries)):
    sum+=summaries[i]+os.linesep+os.linesep
    x=unmasker(summaries[i]+" This is talking about [MASK].")
    z=""
    for y in x:
      z+=y['token_str']+' '
    clas+=z+os.linesep+os.linesep

  return data, silhplot, clustplot, sum, clas
```

```python
import gradio as gr

theme = gr.themes.Soft(
    primary_hue="sky",
    secondary_hue="sky",
)

with gr.Blocks(theme=theme) as Interface:
  gr.Markdown("Upload excel file to perform semantic clustering")
  with gr.Row():
    with gr.Column():
      file = gr.File(file_count='single', file_types=['.xlsx'], type='filepath',scale=1)
      dataset = gr.Dataframe(label='Uploaded dataset',scale=1)
      with gr.Row():
        clusters = gr.Image(label="Spectral Clustering",interactive=False,scale=1)
        sil=gr.Image(label="Silhouette Score Plot",interactive=False,scale=1)
    with gr.Column():
      summaries=gr.Text(label="Cluster Summaries",lines=10,scale=2)
      classes=gr.Text(label="Inferred Classes",lines=10,scale=2)
      btn = gr.Button("Execute",scale=1)

    btn.click(main_function, inputs=file, outputs=[dataset,sil,clusters,summaries,classes])

Interface.launch(share=True,debug=True)
```
